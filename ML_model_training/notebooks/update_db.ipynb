```python
from pymongo import MongoClient
import os 
from dotenv import load_dotenv
import pandas as pd
```

```python
load_dotenv()
```

```python
db_url=os.getenv("DB_URL")
```

```python
print(db_url)
```

```python
client=MongoClient(db_url)
db_name=client["Toyota_cars_2"]
collection=db_name["listings"]
```

```python
condition_map = {
    7: "A+",
    6: "A",
    5: "B+",
    4: "B",
    3: "C+",
    2: "C",
    1: "D"
}
```

```python
df=pd.read_csv("/home/subhan/All/Car_Project/FindACar/ML_model_training/notebooks/updated_Toyota_results.csv")
```

```python
df["condition_label"] = df["condition"].map(condition_map)
```

```python
df.head()
```

```python
for _, row in df.iterrows():
    pk = int(row["primary_key"])  # Ya jo bhi aapka primary key column name hai
    condition = int(row["condition"])  # Ya "condition" column
    label = condition_map.get(condition, "Unknown")
    
    collection.update_one(
        {"primary_key": pk},
        {"$set": {"condition_label": label}}
    )
    
    print(f"âœ… Updated primary_key {pk} with label: {label}")

print("ðŸŽ‰ Done!")
```

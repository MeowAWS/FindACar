```python
import joblib

model = joblib.load("/home/subhan/All/Car_Project/FindACar/ML_model_training/notebooks/model_trained_result.pkl")
le = joblib.load("/home/subhan/All/Car_Project/FindACar/ML_model_training/notebooks/label_encoder.pkl")
```
